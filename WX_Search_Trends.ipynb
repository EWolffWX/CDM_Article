```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
# Figure set up
fig = plt.figure(figsize=(16,10))
fig.patch.set_facecolor('white')

#plt.suptitle('Public Increasingly Turning to YouTube for Weather Information', fontsize=22)

# First plot
# Access data
df_p1 = pd.read_csv('Youtube_Trend_Weather_Forecast.csv')

yt_searches = []
for i in df_p1['Category: All categories'][1:].values:
    if i != '<1':
        num = int(i)
    else:
        num = 0
        
    yt_searches.append(num)
    
yt_searches_ds = pd.Series(yt_searches)

window_size = 12
windows = yt_searches_ds.rolling(window_size, center=True)
yt_moving_averages = windows.mean()

# Plot
ax1 = plt.subplot(221)
x = df_p1.index[1:]
x = np.asarray(x, dtype='datetime64[s]')

ax1.plot(x, yt_searches, linewidth=2, c='tab:blue')
ax1.plot(x, yt_moving_averages, linewidth=1.5, c='k', linestyle='--', label='12-Month Rolling Average')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#plt.xlabel('Date', fontsize=15)
plt.ylabel('Relative Search Interest', fontsize=15)
plt.title('"Weather Forecast"', fontsize=16)
plt.legend(fontsize='large')

plt.grid()

# Second plot
# Access data
df_p2 = pd.read_csv('Youtube_Trend_Live_Weather.csv')

yt_searches = []
for i in df_p2['Category: All categories'][1:].values:
    if i != '<1':
        num = int(i)
    else:
        num = 0
        
    yt_searches.append(num)
    
yt_searches_ds = pd.Series(yt_searches)

window_size = 12
windows = yt_searches_ds.rolling(window_size, center=True)
yt_moving_averages = windows.mean()

# Plot
ax2 = plt.subplot(222)
x = df_p2.index[1:]
x = np.asarray(x, dtype='datetime64[s]')

ax2.plot(x, yt_searches, linewidth=2, c='tab:orange', label='Search Interest')
ax2.plot(x, yt_moving_averages, linewidth=1.5, c='k', linestyle='--', label='12-Month Rolling Average')
ax2.text(17600, 88, 'Hurricanes Irma \nand Maria (2017)', fontsize=12, fontstyle='italic')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

#plt.xlabel('Date', fontsize=15)
#plt.ylabel('Relative Search Interest', fontsize=15)
plt.title('"Live Weather"', fontsize=16)
#plt.legend(fontsize='large')

plt.grid()

# Third plot
# Access data
df_p3 = pd.read_csv('Youtube_Trend_Tornado_Weather.csv')

yt_searches = []
for i in df_p3['Category: All categories'][1:].values:
    if i != '<1':
        num = int(i)
    else:
        num = 0
        
    yt_searches.append(num)
    
yt_searches_ds = pd.Series(yt_searches)

window_size = 12
windows = yt_searches_ds.rolling(window_size, center=True)
yt_moving_averages = windows.mean()

# Plot
ax2 = plt.subplot(223)
x = df_p3.index[1:]
x = np.asarray(x, dtype='datetime64[s]')

ax2.plot(x, yt_searches, linewidth=2, c='tab:red', label='Search Interest')
ax2.plot(x, yt_moving_averages, linewidth=1.5, c='k', linestyle='--', label='12-Month Rolling Average')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Date', fontsize=15)
plt.ylabel('Relative Search Interest', fontsize=15)
plt.title('"Tornado Weather"', fontsize=16)
#plt.legend(fontsize='large')

plt.grid()

# Fourth plot
# Access data
df_p4 = pd.read_csv('Youtube_Trend_Hurricane_Weather.csv')

yt_searches = []
for i in df_p4['Category: All categories'][1:].values:
    if i != '<1':
        num = int(i)
    else:
        num = 0
        
    yt_searches.append(num)
    
yt_searches_ds = pd.Series(yt_searches)

window_size = 12
windows = yt_searches_ds.rolling(window_size, center=True)
yt_moving_averages = windows.mean()

# Plot
ax2 = plt.subplot(224)
x = df_p4.index[1:]
x = np.asarray(x, dtype='datetime64[s]')

ax2.plot(x, yt_searches, linewidth=2, c='tab:green', label='Search Interest')
ax2.plot(x, yt_moving_averages, linewidth=1.5, c='k', linestyle='--', label='12-Month Rolling Average')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Date', fontsize=15)
#plt.ylabel('Relative Search Interest', fontsize=15)
plt.title('"Hurricane Weather"', fontsize=16)
#plt.legend(fontsize='large')

plt.grid()

#plt.savefig('BAMS_CDM_Fig1.png', dpi=300, bbox_inches='tight')
```
